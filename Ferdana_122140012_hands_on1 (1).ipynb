### 1. Membuat Gelombang Sinus dan Cosinus:

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat array waktu t dari 0 hingga 2 dengan langkah 0.0001
t = np.arange(0, 2, 0.0001)

# Membuat sinyal-sinyal sesuai dengan instruksi
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi/4)
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi/2)
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)

# Melihat bentuk array dan nilai-nilai awal
print(f"Bentuk array t: {t.shape}")
print(f"Nilai-nilai awal t: {t[:5]}")
print(f"Nilai-nilai awal y1: {y1[:5]}")
print(f"Nilai-nilai awal y2: {y2[:5]}")
print(f"Nilai-nilai awal y3: {y3[:5]}")
print(f"Nilai-nilai awal y4: {y4[:5]}")
```

### 2. Perbandingan Subplot:

```python
# Membuat figure dengan 2x2 subplot
plt.figure(figsize=(12, 10))

# Subplot 1: y1
plt.subplot(2, 2, 1)
plt.plot(t, y1)
plt.title('$y1 = 2 \cdot \sin(2\pi \cdot 3 \cdot t + 0)$')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Subplot 2: y2
plt.subplot(2, 2, 2)
plt.plot(t, y2)
plt.title('$y2 = 1 \cdot \cos(2\pi \cdot 4 \cdot t + \pi/4)$')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Subplot 3: y3
plt.subplot(2, 2, 3)
plt.plot(t, y3)
plt.title('$y3 = -1 \cdot \sin(2\pi \cdot 5 \cdot t + \pi/2)$')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Subplot 4: y4
plt.subplot(2, 2, 4)
plt.plot(t, y4)
plt.title('$y4 = 0.5 \cdot \cos(2\pi \cdot 6 \cdot t + \pi)$')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

###Analisis

#### 1. Berapa amplitudo dan frekuensi masing-masing sinyal?

Amplitudo masing-masing sinyal:
- y1: 2 (faktor pengali dari sin)
- y2: 1 (faktor pengali dari cos)
- y3: -1 (faktor pengali dari sin, negatif berarti fase terbalik 180 derajat)
- y4: 0.5 (faktor pengali dari cos)

Frekuensi masing-masing sinyal:
- y1: 3 Hz (nilai pengali dari t dalam sin)
- y2: 4 Hz (nilai pengali dari t dalam cos)
- y3: 5 Hz (nilai pengali dari t dalam sin)
- y4: 6 Hz (nilai pengali dari t dalam cos)


#### 2. Bagaimana pergeseran fase mempengaruhi posisi gelombang?

```python
# Membuat contoh sinyal dengan fase berbeda untuk demonstrasi
plt.figure(figsize=(12, 6))

# Frekuensi dan waktu yang sama
f = 1
t_demo = np.arange(0, 2, 0.001)

# Sinyal dengan fase berbeda
s1 = np.sin(2 * np.pi * f * t_demo + 0)
s2 = np.sin(2 * np.pi * f * t_demo + np.pi/4)  # Pergeseran 45 derajat
s3 = np.sin(2 * np.pi * f * t_demo + np.pi/2)  # Pergeseran 90 derajat
s4 = np.sin(2 * np.pi * f * t_demo + np.pi)    # Pergeseran 180 derajat

plt.plot(t_demo, s1, label='Fase 0')
plt.plot(t_demo, s2, label='Fase π/4 (45°)')
plt.plot(t_demo, s3, label='Fase π/2 (90°)')
plt.plot(t_demo, s4, label='Fase π (180°)')

plt.title('Pengaruh Pergeseran Fase pada Sinyal Sinus')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

Pergeseran fase mempengaruhi posisi gelombang sebagai berikut:
- Fase 0: Sinyal dimulai dari nilai 0 dan naik (sinyal sin klasik)
- Fase π/4 (45°): Sinyal bergeser ke kiri sebesar 1/8 periode
- Fase π/2 (90°): Sinyal bergeser ke kiri sebesar 1/4 periode (bentuknya seperti cos)
- Fase π (180°): Sinyal terbalik (fase bergeser ke kiri sebesar 1/2 periode)

Pada tugas, kita melihat:
- y1 dengan fase 0: Sinyal sinus standar
- y2 dengan fase π/4: Sinyal cosinus bergeser 45 derajat ke kiri
- y3 dengan fase π/2: Sinyal sinus bergeser 90 derajat ke kiri (menjadi -cos)
- y4 dengan fase π: Sinyal cosinus bergeser 180 derajat (terbalik)

#### 3. Bandingkan sinyal-sinyal dengan amplitudo yang berbeda

```python
# Membuat perbandingan amplitudo
plt.figure(figsize=(12, 6))

t_demo = np.arange(0, 2, 0.001)
f = 1

# Sinyal dengan amplitudo berbeda
a1 = 0.5 * np.sin(2 * np.pi * f * t_demo)
a2 = 1.0 * np.sin(2 * np.pi * f * t_demo)
a3 = 2.0 * np.sin(2 * np.pi * f * t_demo)

plt.plot(t_demo, a1, label='Amplitudo 0.5')
plt.plot(t_demo, a2, label='Amplitudo 1.0')
plt.plot(t_demo, a3, label='Amplitudo 2.0')

plt.title('Pengaruh Amplitudo pada Sinyal Sinus')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

Amplitudo mempengaruhi tampilan gelombang sebagai berikut:
- Amplitudo mengatur tinggi maksimum (puncak) dan kedalaman minimum (lembah) dari sinyal.
- Semakin besar amplitudo, semakin tinggi puncak dan semakin dalam lembahnya.
- Frekuensi dan bentuk gelombang tetap sama, hanya skala vertikal yang berubah.

Dari sinyal-sinyal dalam tugas:
- y1 (amplitudo 2): Memiliki puncak tertinggi dan lembah terdalam
- y2 (amplitudo 1): Memiliki puncak dan lembah standar
- y3 (amplitudo -1): Memiliki puncak dan lembah standar tetapi terbalik (fase bergeser 180°)
- y4 (amplitudo 0.5): Memiliki puncak dan lembah paling rendah

#### 4. Bandingkan sinyal-sinyal dengan pergeseran fase yang berbeda

```python
# Membuat perbandingan fase pada sinyal cosinus
plt.figure(figsize=(12, 6))

t_demo = np.arange(0, 2, 0.001)
f = 1

# Sinyal cosinus dengan fase berbeda
p1 = np.cos(2 * np.pi * f * t_demo + 0)
p2 = np.cos(2 * np.pi * f * t_demo + np.pi/4)
p3 = np.cos(2 * np.pi * f * t_demo + np.pi/2)
p4 = np.cos(2 * np.pi * f * t_demo + np.pi)

plt.plot(t_demo, p1, label='Fase 0')
plt.plot(t_demo, p2, label='Fase π/4 (45°)')
plt.plot(t_demo, p3, label='Fase π/2 (90°)')
plt.plot(t_demo, p4, label='Fase π (180°)')

plt.title('Pengaruh Pergeseran Fase pada Sinyal Cosinus')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

Pergeseran fase mempengaruhi tampilan gelombang sebagai berikut:
- Fase 0 pada cosinus: Sinyal dimulai dari nilai 1 (maksimum)
- Fase π/4 (45°): Sinyal bergeser ke kiri sebesar 1/8 periode
- Fase π/2 (90°): Sinyal bergeser ke kiri sebesar 1/4 periode (bentuknya seperti -sin)
- Fase π (180°): Sinyal terbalik (fase bergeser ke kiri sebesar 1/2 periode)

Pada sinyal-sinyal dalam tugas:
- y1 (sin dengan fase 0): Bentuk standar gelombang sinus
- y2 (cos dengan fase π/4): Cosinus bergeser 45° ke kiri
- y3 (sin dengan fase π/2): Sinus bergeser 90° ke kiri (mirip -cos)
- y4 (cos dengan fase π): Cosinus bergeser 180° ke kiri (terbalik)

### 4. Tugas Lanjutan:

```python
# Membuat sinyal baru y5 sebagai kombinasi dari y1 dan y2
y5 = y1 + y2

# Plot y5 dan diskusikan hasilnya
plt.figure(figsize=(12, 6))

# Plot sinyal asli
plt.plot(t, y1, label='y1 = 2·sin(2π·3·t)')
plt.plot(t, y2, label='y2 = 1·cos(2π·4·t + π/4)')
plt.plot(t, y5, label='y5 = y1 + y2', linewidth=2)

plt.title('Kombinasi Sinyal Sinus dan Cosinus')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.xlim(0, 0.5)  # Memperbesar tampilan untuk melihat detail

plt.show()
```

#### Diskusi Sinyal Kombinasi

Kombinasi dua gelombang sinus/cosinus dengan frekuensi berbeda (y1 dengan frekuensi 3 Hz dan y2 dengan frekuensi 4 Hz) menghasilkan bentuk gelombang yang lebih kompleks dengan karakteristik berikut:

1. Pola Interferensi: Sinyal y5 menunjukkan pola interferensi antara dua frekuensi berbeda.
   - Terjadi interferensi konstruktif ketika kedua sinyal mencapai puncak bersama, menghasilkan amplitudo gabungan maksimum sekitar 3.
   - Terjadi interferensi destruktif ketika satu sinyal berada di puncak dan yang lain di lembah, mengurangi amplitudo gabungan.

2. Modifikasi Amplitudo: Sinyal gabungan memiliki amplitudo bervariasi yang disebut "beating" atau fenomena pelayangan.
   - Pelayangan ini terjadi pada frekuensi yang merupakan selisih dari kedua frekuensi komponen (dalam kasus ini, 4 Hz - 3 Hz = 1 Hz).

3. Bentuk Gelombang Tidak Reguler: Meskipun kedua komponen sinyal adalah periodik, sinyal gabungan akan menunjukkan pola yang lebih kompleks.
   - Periodikas total dari sinyal gabungan adalah kelipatan persekutuan terkecil dari periode kedua sinyal asli.

4. Sinyal y5 juga menunjukkan superposisi linier dari kedua komponen - prinsip penting dalam pengolahan sinyal yang menyatakan bahwa sinyal gabungan adalah jumlah dari masing-masing komponen individu.

Fenomena ini sangat penting dalam berbagai aplikasi seperti akustik (menghasilkan nada beat), komunikasi (modulasi sinyal), dan analisis spektral (dekomposisi sinyal kompleks).
